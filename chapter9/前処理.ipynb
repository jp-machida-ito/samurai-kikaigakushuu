```python
#前処理


#データの前処理
#1.データの取り込み　2.データの概要理解　3.データの前処理　4.前処理済みデータの保存

#1.データの取り込み
import pandas as pd
import seaborn as sns

#カリフォルニアの住宅価格データセットを読み込む
from sklearn.datasets import fetch_california_housing
dataset = fetch_california_housing()
print(dataset)
```

```python
df =pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df['price'] = dataset.target
df.head()
```

```python
print(dataset.DESCR)
```

```python
feature_names_JPN = ['所得','築年数','部屋数','寝室数','地域人口','世帯人数','緯度','経度','住宅価格']
df.columns = feature_names_JPN
df.head()
```

```python
df.shape #データの行数と列数を確認
```

```python
df.info() #データの型を確認 non-nullは欠損値の数を表す
```

```python
df.describe() #データの統計量を確認
```

```python
import japanize_matplotlib
df.hist(bins=30,figsize=(12,10)) #ヒストグラムを描画

#pandasのdf.hist()は、裏側でmatplotlibを使用してヒストグラムを描画しています。
#グラフを微調整したい、画像として保存したい等の状況になったら、matplotlibを使用しましょう！

#そして、データの大まかな特徴を確認していきます。(外れ値の有無、分布の偏り等)
```

```python
#ヒストグラムを確認した結果、地域人口に関して大きい値の分布を把握しにくい状況です。
#そこで、seabornを使用し、データの分布の可視化をしましょう！

sns.stripplot(x='地域人口', data=df)
#seabornのstripplotは、データの分布を点で表現するグラフです。  
#このグラフを確認すると、ボリュームゾーンから大きく離れた値がいくつか存在することがわかります。
#ただ今回は、外れ値を削除せずに分析を進めていきます。
```

```python
#築年数も同様に、分布状況を確認してみましょう！
#最大値が大きく離れていることがわかります。

#df.describe()で確認した結果、築年数の最大値が大きく離れていることがわかります。
#dfから築年数が52であるデータを抽出し、shapeで行数を確認してみましょう。
df.query('築年数 == 52').shape
```

```python
#築年数が52年のデータが1000件以上あるとは考えにくいですね。
#国勢調査で築年数52以上のデータは一律52とされている可能性があります。
#したがって、築年数が52のデータは不適切なデータの可能性が高いと判断し、除外しましょう！

df = df[df['築年数'] != 52] 
```

```python
#住宅価格
df.query('住宅価格 == 5.000010').shape
```

```python
df = df[df['住宅価格'] != 5.000010]
```

```python
#部屋数、寝室数、世帯人数
#現実的に考えにくいデータが存在しています・・
#正しい情報への修正は困難な為、部屋数・寝室数・世帯人数の情報は、データ分析に利用しない方針で進めていきましょう！
#代わりに、部屋数/人、寝室数/人という情報を新たに作成してみましょう！

#世帯数の作成
df['世帯数'] = df['地域人口']/df['世帯人数']
df.head()
```

```python
#部屋数（1世帯当たりの平均部屋数）に世帯数をかけて、全部屋数の列を新たに作成！
#寝室数も同様に、1世帯当たりの平均寝室数に世帯数をかけて、全部屋数の列を新たに作成してみましょう！

df['全部屋数'] = df['部屋数'] * df['世帯数']
df['全部寝室数'] = df['寝室数'] * df['世帯数']
df.head()
```

```python
#全部屋数を地域人口で割って、部屋数/人の列を新たに作成してみましょう！
#全部寝室数も同様に、地域人口で割って、寝室数/人の列を新たに作成してみましょう！

df['部屋数/人'] = df['全部屋数'] / df['地域人口']
df['寝室数/人'] = df['全部寝室数'] / df['地域人口']
df.head()
```

```python
#この情報が果たして使用できるのか・・
#seabornのstripplotで、部屋数/人と寝室数/人の分布を確認してみましょう！
sns.scatterplot(x='部屋数/人', y='寝室数/人', data=df)

#これは十分に使用できる情報ですね！
```

```python
#現実的に考えにくい部屋数・寝室数・世帯人数の情報は削除しましょう
#また、部屋数/人と寝室数/人の情報作成の元となった、全部屋数と全部寝室数の情報も削除しましょう！

df = df.drop(columns=['部屋数','寝室数','世帯人数','全部屋数','全部寝室数','世帯数'])
df.head()
```

```python
#改めてヒストグラムで現状を確認しましょう！
df.hist(bins=30,figsize=(12,10))
```

```python
#それでは、ここまでの情報を前処理済みデータとして保存していきましょう！
df.to_csv('california_housing_cleansing.csv',index=False,encoding='utf-8-sig')
```
